1. Difference Between a Function and a Method in Python
 - Function: A standalone block of code defined using the def keyword. It
   operates on data passed as arguments.
 - Method: A function associated with an object. It operates on the object  
   itself and is called using the dot (.) notation.
2. Concept of Function Arguments and Parameters in Python
 - Parameters: Variables in a function definition used to accept input values.
 - Arguments: Actual values passed to the function when it is called.
3. Ways to Define and Call a Function in Python
 - Define: Use the def keyword. Optionally use lambda for anonymous functions.
 - Call: Type the name of the function and enter the arguments within the   
   paranthesis
  Example:
      def square(x):  # Define
        return x ** 2

      print(square(5))  # Call
4. Purpose of the return Statement in a Python Function
 - The return statement sends a value back to the caller and exits the function.
   Without return, the function returns None by default
5. Iterators vs. Iterables
 - Iterable: An object capable of returning one element at a time (e.g., lists, tuples, strings).
 - Iterator: An object representing a stream of data from an iterable, created using iter(). It produces elements with next().
6. Generators in Python
 - Special functions that yield values one at a time using yield instead of return.
 - Automatically create an iterator.
  Example:
  def count_up_to(n):
      for i in range(1, n + 1):
          yield i

  for num in count_up_to(3):
      print(num)  # Output: 1, 2, 3
7. Advantages of Generators Over Regular Functions
 - Memory Efficient: Generate items on demand instead of storing them in memory.
 - Infinite Sequences: Suitable for representing streams of data or infinite sequences.
 - Simpler Syntax: Easy to implement lazy evaluation.
8. Lambda Function in Python
 - Anonymous, inline functions defined using the lambda keyword.
 - Typically used for short, one-time operations
  Example:
  square = lambda x: x ** 2
  print(square(4))  # Output: 16
9. Purpose and Usage of map() Function
 - Applies a given function to all items in an iterable and returns a map   
   object
   Example:
   nums = [1, 2, 3]
  squares = map(lambda x: x ** 2, nums)
  print(list(squares))  # Output: [1, 4, 9]
10. Difference Between map(), reduce(), and filter()
 - map(): Transforms each element of an iterable using a function.
  Example:
  map(lambda x: x * 2, [1, 2, 3])  # Output: [2, 4, 6]
 - reduce(): Applies a function cumulatively to items of an iterable (requires  
   functools)
   Example:
   from functools import reduce
   reduce(lambda x, y: x + y, [1, 2, 3])  # Output: 6
 - filter(): Filters elements based on a condition.
   Example:
   filter(lambda x: x > 1, [1, 2, 3])  # Output: [2, 3]
11. Using pen & Paper write the internal mechanism for sum operation using    
    reduce function on this given list:[47,11,42,13];
 - Mechanism: The reduce() function starts by taking the first two elements of the list, applies the function, and then replaces them with the result.
It continues by applying the function to the result and the next element in the list until one value remains.
 Example:
  from functools import reduce
  nums = [47, 11, 42, 13]
  result = reduce(lambda x, y: x + y, nums)
  print(result)  # Output: 113


```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
# the list
def sum(l:list)->int:
    total = 0
    for i in l:
        if i%2==0:
            total += i
    return total
print(sum([3, 5, 1, 4 ,2, 5, 3, 4]))
```

```python
# 2. Create a Python function that accepts a string and returns the reverse of that string.
def reversed(st: str)->str:
    reverse_st = ""
    for i in range((len(st)-1), -1, -1):
        reverse_st += st[i]
    return reverse_st
n = str(input("Enter a string: "))
print(reversed(n))
```

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
# each number.
def sqaured(l:list)->list:
    new_list = []
    for i in l:
        new_list.append(i**2)
    return new_list
print(sqaured([4, 2, 4, 1, 5, 7, 5, 2]))
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def check_prime():
    for i in range(200):
        primecheck = True
        if i%2 == 0:
            print(f"{i} is not a prime number")
            continue
        for l in range(1,int(i/2),2):
            if l == 0 or l == 1 or l == 2:
                continue
            if i%l==0:
                primecheck= False
                break
        if primecheck == True:
            print(f"{i} is a prime number")
        else:
            print(f"{i} is not a prime number")
check_prime()
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms.
class fibo:
    def fibonacci(self):
        a, b = 0, 1
        while True:
            yield a
            a, b = b, a+b
l = fibo()

f = iter(l.fibonacci())
for i in range(10):
    print(f"{next(f)} ", end="")
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power():
    a = 2
    while True:
        yield a
        a *=2
range_of_print = int(input("Enter the range: "))
f = iter(power())
for i in range(range_of_print):
    print(next(f))
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read(fi):
        with open(fi, 'r') as f:
            while True:
                yield f.readline()
file_read = iter(read("file.txt"))
for i in range(4):
    st = str(next(file_read))
    print(st,end="")
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
l = [(3,5), (1,2), (5,4), (2,1)]
sorted_list = sorted(l, key=lambda x: x[1])
print("Before sorting:", l)
print("After sorting:", sorted_list)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius_list = [23, 35 ,12, 43]
fahrenhiet_list = list(map(lambda x: x*9/5+32, celsius_list))
print(fahrenhiet_list)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
st = "Hello, My name is Steve"
new_st = ''.join(filter(lambda x: False if x in 'aioue' or x in 'AEIOU' else True, st))
print(new_st)
```
